```python
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

#out_avi = cv2.VideoWriter('final avira vid.mp4',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))
out_mp4 = cv2.VideoWriter('final avira vid.mp4',cv2.VideoWriter_fourcc(*'XVID'), 10, (frame_width,frame_height))
```

```python
while(cap.isOpened()):
    
    ret, frame = cap.read()
    if ret == True:
        #out_avi.write(frame)
        out_mp4.write(frame)
        
    else:
        break
```

```python
cap.release()
#out_avi.release()
out_mp4.release()
```
